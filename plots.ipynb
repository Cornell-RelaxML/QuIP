```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_theme()
```

```python
params = [5.71,	22.05,	86.56]
ViT_data = {
     'Full Precision': [0.7546,0.8139,	0.851],
	'GPTQ': [0.002,	0.004,	0.2844],
	'QuIP': [0.0142,0.2198,	0.7754],
	'QuIP 2sigma': [0.091,	0.4896,	0.7954],
	'FrameQuant': [0.0892,0.4955,	0.7971],
	'FrameQuant 1.1': [0.2579,0.6133,	0.8087], 
}
```

```python
from itertools import product
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
for i,j in product((0,1), (0,1)):
    for k,v in ViT_data.items():
        axes[i][j].semilogx(params, v)
    axes[i][j].legend(list(ViT_data.keys()))
plt.tight_layout()
plt.savefig('placeholder.png')
plt.show()
```
