```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
ViT_params = [5.71,	22.05,	38.861,	86.56,	304.32,	632.04]
vit_fp = [0.7546, 0.8139, 0.8445, 0.8510, 0.8584, 0.8759]
vit_no_pl = [0.1418, 0.5069, 0.7639, 0.8049, 0.8321, 0.8572]
vit_no_pl_1p1 = [0.2797, 0.6018, 0.7798, 0.8092, 0.8384, 0.8587]
vit_no_pl_1p2 = [0.3204, 0.6450, 0.7914, 0.8127, 0.8385, 0.8594]
vit_no_pl_1p3 = [0.4291, 0.6698, 0.7959, 0.8159, 0.8383, 0.8565]
vit_with_pl = [0.0909, 0.4905, 0.7564, 0.7976, 0.8321, 0.8559]
# vit_with_pl = [0.1018,	0.481,	0.7581,	0.7953,	0.8322,	0.8549]
```

```python
plt.semilogx(ViT_params, [i*100 for i in vit_fp], label='Full Presion', marker='o', linestyle='-')
plt.semilogx(ViT_params, [i*100 for i in vit_no_pl], label='FrameQuant without P_l, r=1', marker='s', linestyle='--')
plt.semilogx(ViT_params, [i*100 for i in vit_no_pl_1p1], label='FrameQuant without P_l, r=1.1', marker='+', linestyle=(0, (3, 5, 1, 5)))
plt.semilogx(ViT_params, [i*100 for i in vit_no_pl_1p2], label='FrameQuant without P_l, r=1.2', marker='X', linestyle=(0, (1, 1)))
plt.semilogx(ViT_params, [i*100 for i in vit_no_pl_1p3], label='FrameQuant without P_l, r=1.3', marker='D', linestyle=(0, (3, 1, 1, 1, 1, 1)))
plt.semilogx(ViT_params, [i*100 for i in vit_with_pl], label='FrameQuant with P_l, r=1', marker='^', linestyle=':')
plt.legend()
plt.xlabel('Number of parameters in Millions')
plt.ylabel('ImageNet-1k accuracy')
plt.title('ViT models')
plt.ylim([-2, 90])
plt.show()
```
